K MEANS CLUSTERING is an unsupervised learning algorithm that will attempt to group similar clusters together in your data. Which means, it takes unlabeled data.

The overall goal is to divide data into distinct groups such that observations within each group are similar

# The K MEANS Algorithm

* Choose a number of Clusters "K"
* Randomly assign each point to a cluster
* Until clusters stop changing, repeat the following:
    * For each cluster, compute the cluster centroid bu taking the mean vector of points in the cluster
    * Assign each data point to the cluster for which the centroid is the closest

# Choosig a K Value

* There is no easy answer for choosing a "best" K value
* One way is the elbow method:
    * First of all, compute the sum of squared erro for some values of K
    * The SSE is defined as the sum of the squared distance between each member of the cluster and its centroid
    * The error decreases as K gets larger. As the number of clusters increases, the distorsion is smaller
    * The idea of the Elbow method is to choose the K where the SSE decreases abruptly

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set(style="darkgrid")
```

```python
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples=200,n_features=2,centers = 4,cluster_std=1.8, 
                  random_state=101)
```

```python
data[0].shape
```

```python
plt.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
fig , (ax1,ax2) = plt.subplots(1,2,sharey=True,figsize=(10,6))

ax1.set_title('K Means')
ax1.scatter(data[0][:,0],data[0][:,1],c=kmeans.labels_,cmap='rainbow')

ax2.set_title('Original')
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```
